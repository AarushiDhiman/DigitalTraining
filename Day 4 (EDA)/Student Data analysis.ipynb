```python
import pandas as pd
#Create student dataframe by reading csv file
df=pd.read_csv('Marks_data.csv')
df.head()  # checking first 5 rows
```

```python
df.isna()
#df.isnull()
```

```python
#Total marks for each student
df['Total']=df[['Math','Science','English']].sum(axis=1)
df.head(5)
```

```python
# Verifying the Total
df[['Name','Math','Science','English','Total']].head()
```

```python
# Function to Assign grades
def assign_grade(Total):
    if Total>=250:
        return 'A'
    elif Total>=200:
        return 'B'
    elif Total>=150:
        return 'C'
    elif Total>=100:
        return 'D'
    else:
        return 'F'
```

```python
#Using the Function to generate grades in student_data
df['Grade']=df['Total'].apply(assign_grade)

print(df[['Name','Total','Grade']].head(11))
```

```python
#Filter the data to find out students who got grade A

grade_A_student=df[df['Grade']=='A']
print(grade_A_student)
```

```python
#Find total student failed
grade_f_student = df[df['Grade']=='F']
print("Total student Failed: ", len(grade_f_student))
```

```python
#Find out the maximum marks in subject maths
math_max=max(df['Math'])
print("Maximum marks in maths is:",math_max)

print(" ")
print("Student data who got maximum in Maths: ")
print(df[df['Math']==math_max])
```

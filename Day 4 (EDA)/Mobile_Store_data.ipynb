```python
import pandas as pd
```

```python
# Reading the csv file
df=pd.read_csv('Mobile_Store_data.csv')
df.head(5)
```

```python
#Checking null Value
df.isnull()
```

```python
df['Total_Amount']=df['Quantity']*df['Price']
df.head(20)
```

```python
df=df.drop(columns='Total_amount')
```

```python
most_sold=max(df['Quantity'])
print("Most Sold Product Quantity:",most_sold)
print(df[df['Quantity']==most_sold])
```

```python
df.head()
```

```python
#Maximum sold using group by
most_sold_product= df.groupby('Product')['Quantity'].sum()
max_sold= most_sold_product.max()
print("Maximum sold product: ",df[df['Quantity']==most_sold])
```

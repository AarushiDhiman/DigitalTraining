```python
import matplotlib.pyplot as plt
print(plt.__package__)  #See the package
```

```python

print("--------Simple line chart----------")
plt.figure(figsize=(6,4)) 

plt.plot([1,2,3],[4,5,6]) 
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,25,30,40]

plt.plot(x,y, marker='o', linestyle='-',color='red')

plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.title("Scatter Plot")
plt.grid(True)
plt.show()
```

```python
x=[11,15,18,25,35]
y=[10,20,25,30,40]

plt.scatter(x,y, marker='*', color='red')

plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.title("Scatter Plot")
plt.grid(True)
plt.show()
```

```python

categories = ['A','B','C','D']
values = [5,7,3,8]

plt.bar(categories,values,color='lavender')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

```python
import numpy as np
data=np.random.randn(1000)
plt.hist(data,bins=30,color='black',edgecolor='pink')
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
sizes=[35,30,25,15]
labels=['JS','DSA','ANGULAR','PYTHON']
colors=['pink','skyblue','lavender','yellow']
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Creating Data
sales=np.array([1300,2500,6000,2300,1900,1800,2100,1750,2100,3900])
days=np.arange(1,11)  # array of 1 to 10

#printing sales data
print("Sales data:",sales)
```

```python
print("-----------Basic Analysis--------")
total_sales=np.sum(sales)
average_sales=np.mean(sales)
highest_sales=np.max(sales)
lowest_sales=np.mean(sales)

print(f"Total_sales: {total_sales}")
print(f"Average_sales: {average_sales}")
print(f"Highest_sales: {highest_sales}")
print(f"Lowest_sales: {lowest_sales}")
```

```python
#LINE CHART- SALES TREND OVER DAYS
plt.figure(figsize=(8,5))
plt.plot(days,sales,marker='*',linestyle='-',color='red',label='Sales')
plt.axhline(average_sales,linestyle='--',label='Average_Sales')

plt.xlabel("Days")
plt.ylabel("Sales Revenue ($)")
plt.title("Sales Trend over 10 Days")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# BAR CHART - SALES COMPARISON BY DAY
plt.figure(figsize=(8,5))
plt.bar(days, sales,color='c',label='sales')
plt.xlabel("Days")
plt.ylabel("Sales Revenue ($)")
plt.title("Sales per Day")
plt.legend()
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

```python
#PIE CHART-SALES CONTRIBUTION PER DAY
plt.figure(figsize=(6,6))
plt.pie(sales,labels=days,colors=plt.cm.Paired.colors,autopct='%1.1f%%')

plt.title("Sales Contribution Per Day")
plt.show()
```
